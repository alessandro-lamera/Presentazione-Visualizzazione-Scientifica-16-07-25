```python
import pandas as pd
import matplotlib.pyplot as plt

# === CONFIGURAZIONI ===
csv_file = "./Foi_medie_annue_dal_2016.csv" # Nome del file CSV
x_column = "TIME_PERIOD"             # Colonna da usare sull'asse X
y_column = "Osservazione"            # Colonna da usare sull'asse Y
graph_title = "INFLAZIONE"
y_label = "inflazione (base 2015 =100)"
x_label = "Anni"
output_file = "grafico.png"    # Nome immagine da salvare

# === LEGGI IL CSV ===
try:
    df = pd.read_csv("./Foi_medie_annue_dal_2016.csv", delimiter=",")

    # === CREA IL GRAFICO ===
        # === CREA IL GRAFICO ===
    plt.figure(figsize=(10, 6))
    plt.plot(df[x_column], df[y_column], marker='o', linestyle='-', color='royalblue', linewidth=2)

    # === RIMUOVI BORDI DESTRO E SUPERIORE ===
    ax = plt.gca()  # Ottieni l'oggetto degli assi correnti
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # === STILE ===
    plt.title(graph_title, fontsize=16, fontweight='bold')
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel(y_label, fontsize=12)
   # plt.grid(True, linestyle='--', alpha=0.6)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # === SALVA E MOSTRA ===
    plt.savefig(output_file, dpi=300)
    plt.show()

except FileNotFoundError:
    print(f" Errore: Il file '{csv_file}' non  stato trovato.")
except KeyError as e:
    print(f" Errore: Colonna non trovata nel CSV: {e}")
```
